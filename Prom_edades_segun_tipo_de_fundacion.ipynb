```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import date, time, datetime
%matplotlib inline
```

```python
train_labels = pd.read_csv('train_labels.csv')
train_labels
```

```python
train_values = pd.read_csv('train_values.csv', usecols=['building_id', 'age', 'foundation_type'])
train_values.shape
```

```python
merged = train_values.merge(train_labels, how='inner')
merged
```

```python
merged_avg = merged.groupby(['foundation_type', 'damage_grade']).agg({'age':'mean'})
merged_avg.columns = ['age_avg']
merged_avg.reset_index()
```

```python
pivot_table = merged_avg.pivot_table(index='foundation_type', columns='damage_grade', values='age_avg', fill_value='None')
pivot_table
```
